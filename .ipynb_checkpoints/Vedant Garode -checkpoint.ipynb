# Total Impression By Age Group 30 -34

```python
agedata = data[(data["age"] == "30-34") | ((data["age"] >= "30") & (data["age"] <= "34"))]
print("total count of age group :-",agedata["impressions"].count())
print("total Impression Count of Age group 30-34 :-",agedata["impressions"].sum())
```

# All ad_ids for every campaign_id 

```python
iddata = data.groupby("campaign_id")["ad_id"]
```

```python
iddata.get_group("916")
```

# total clicks where report_start between dates 19/08/2017 to 22/08/2017

```python
clickdata = data[(data["reporting_start"] >= "19/08/2017") & (data["reporting_end"] >= "22/08/2017")]["clicks"]
```

```python
print("total clicks where report_start between dates 19/08/2017 to 22/08/2017 :-",clickdata.sum())
```
